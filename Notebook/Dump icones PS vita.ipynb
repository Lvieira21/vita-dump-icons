# Setup

```python
PSV_IP = '192.168.1.22'
PORT = 1337
download_icon_to = "D:/PS Vita/Icones"
```

# Inital steps

```python
import os
from ftplib import FTP #FTP Access
from PIL import Image #resizing
```

```python
APPMETA_VITA = 'ur0:/appmeta'
```

```python
ftp = FTP()
ftp.connect(PSV_IP, PORT)
ftp.login()
```

```python
try:
    os.mkdir(download_icon_to)
except OSError:
    print ("Creation of the directory %s failed" % download_icon_to)
else:
    print ("Successfully created the directory %s " % download_icon_to)
```

# Download Icons - FTP

```python
ftp.cwd(APPMETA_VITA)
```

```python
pastas = []
nomes = []
ftp.dir("", pastas.append)

#Get folder names
for pasta in pastas:
    pasta = pasta.split()
    nomes.append(pasta[8])
```

```python
for nome in nomes:
    
    ftp.cwd(nome)
    try:
        ftp.retrbinary("RETR " + 'icon0.png', open(os.path.join(download_icon_to, nome + '.png'), 'wb').write) #download icon
        print("Game ID{} image downloaded successfully".format(nome))
        imagem = Image.open(os.path.join(download_icon_to, nome + '.png')) #find downloaded image
        imagem = imagem.resize((512,512))
        imagem.save(os.path.join(download_icon_to, nome + '.png')) #overwriting resized image onto original
        print("{} resized".format(nome))
        ftp.cwd('..')
    except:
        print("Game ID {} image was not found".format(nome))
        ftp.cwd('..')
 ftp.close()   
```
